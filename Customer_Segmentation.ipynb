```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Read the CSV file
df = pd.read_csv("Comined_5000.csv")

# Drop unnecessary columns
df = df.drop(['CustomerID', 'PhoneNumber', 'ZipCode'], axis=1)

# Convert PurchaseDate to datetime
df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'])

# Feature selection (you can add more features as per your requirement)
selected_features = ['Age', 'Price', 'Quantity']
X = df[selected_features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform K-means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Create a pivot table with clusters as rows and selected features as columns
pivot_table = df.pivot_table(index='Cluster', values=selected_features, aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap='viridis')
plt.title('Cluster Centers - Heatmap')
plt.xlabel('Selected Features')
plt.ylabel('Cluster')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

# Read the CSV file
df = pd.read_csv("Comined_5000.csv")

# Drop unnecessary columns
df = df.drop(['CustomerID', 'PhoneNumber', 'ZipCode'], axis=1)

# Convert PurchaseDate to datetime
df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'])

# Feature selection (you can add more features as per your requirement)
selected_features = ['Age', 'Price', 'Quantity']
X = df[selected_features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Hyperparameter tuning using the Elbow Method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Determine the optimal number of clusters using Silhouette Score
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_scores.append(metrics.silhouette_score(X_scaled, cluster_labels))

# Plotting Silhouette Score
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()

# Based on the analysis, choose the optimal number of clusters
optimal_k = 3

# Perform K-means clustering with optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Create a pivot table with clusters as rows and selected features as columns
pivot_table = df.pivot_table(index='Cluster', values=selected_features, aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap='viridis')
plt.title('Cluster Centers - Heatmap')
plt.xlabel('Selected Features')
plt.ylabel('Cluster')
plt.show()
```

# Heirarchical Segmentation

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import AgglomerativeClustering

# Load your CSV data
data = pd.read_csv("Comined_5000.csv")

# Select relevant features for clustering
features = ['Age', 'Price', 'Quantity']

# Preprocess the data
X = data[features]

# Encode categorical variables
encoder = OneHotEncoder(drop='first', sparse=False)
gender_encoded = encoder.fit_transform(data[['Gender']])
X_encoded = pd.concat([X, pd.DataFrame(gender_encoded, columns=['Gender'])], axis=1)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Perform hierarchical clustering
model = AgglomerativeClustering(n_clusters=5, linkage='ward')  # You can adjust the number of clusters and linkage method
clusters = model.fit_predict(X_scaled)

# Add cluster labels to the original dataframe
data['Cluster'] = clusters

# Print cluster sizes
print("Cluster sizes:")
print(data['Cluster'].value_counts())

# Print cluster centroids (mean values of each feature within each cluster)
print("\nCluster centroids:")
print(data.groupby('Cluster')[features].mean())
```
